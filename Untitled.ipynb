```python
import xml.etree.ElementTree as ET
from pytube import YouTube
import nltk
```

```python
def get_caption(url):
    yt = YouTube(url)
    caption = yt.captions.get_by_language_code('a.en')
    xml_text=caption.xml_captions
    root = ET.fromstring(xml_text)
    full_text=[]
    for word in root.iter('s'):
        word=word.text
        
        full_text.append(word.text)
    return full_text
```

```python
text=get_caption('https://www.youtube.com/watch?v=76tyRPCRQ-Y')
text
```

```python
for i in text:
    print(i.strip())
```
